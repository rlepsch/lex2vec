# Lex2Vec

O Lex2Vec consiste no treinamento de um modelo [word2vec](https://www.tensorflow.org/tutorials/word2vec) (Continuous Bag-of-Words - CBOW e Skip-Gram) para gerar representações vetoriais de termos presentes no _corpus_ da legislação federal brasileira. Para treinar o modelo foram utilizadas normas federais promulgadas entre 1824 a 2017 (pouco mais de 230 mil normas), extraídas do Sistema de Legislação Informatizada da Câmara dos Deputados.

Este notebook descreve os passos empregados na treinamento do modelo.

### Requisitos de hardware

## Preparando o Corpus

#### Carregando normas do sistema de arquivos

```python
lex_files = sorted(glob.glob("data/legin_xml/fed/*/*/*.xml"))
```

```python
print("Found {} files".format(len(lex_files)))
```

#### Combinando todas as normas em uma string

```python
corpus_raw = u""
invalid_count = 0

for lex_file in tqdm(lex_files):
    
    try:
        tree = ElementTree.parse(lex_file)
        root = tree.getroot()
        corpus_raw += root.find("./corpodocumento").text
        
    except ElementTree.ParseError as err:
        print(err, ': ', lex_file)
        invalid_count += 1
        
print("Corpus is now {0} characters long".format(len(corpus_raw)))
```

É importante salvar o resultado desse pré-processamento, já que ele leva aproximadamente 6 horas!

```python
import io
with io.open("data/corpus.txt", "w", encoding="utf-8") as f:
    f.write(corpus_raw)
```

#### Quebrando o corpus em sentenças

```python
tokenizer = nltk.data.load('tokenizers/punkt/portuguese.pickle')
```

```python
raw_sentences = tokenizer.tokenize(corpus_raw)
```

```python
# converte em uma lista de palavras
# remove acentos e caracteres desnecessários
def sentence_to_wordlist(raw):
    clean = unidecode(raw)
    clean = re.sub("[^a-zA-Z]"," ", clean)
    words = clean.split()
    return words
```

```python
sentences = []
for raw_sentence in tqdm(raw_sentences):
    if len(raw_sentence) > 0:
        sentences.append(sentence_to_wordlist(raw_sentence))
```

Examinando uma amostra de sentença:

```python
choice = np.random.choice(range(len(raw_sentences)))
print(raw_sentences[choice])
print(sentence_to_wordlist(raw_sentences[choice]))
```

```python
import pickle
pickle.dump(raw_sentences, open("data/raw_sentences.p", "wb"))
pickle.dump(sentences, open("data/sentences.p", "wb"))
```

```python
token_count = sum([len(sentence) for sentence in sentences])
print("The corpus contains {0:,} tokens".format(token_count))
```

## Treinando o modelo Word2Vec

```python
# Dimensionalidade dos word vectors.
# more dimensions = more generalized
num_features = 300

# Minimum word count threshold.
min_word_count = 50

# Number of threads to run in parallel.
# more workers, faster we train
num_workers = multiprocessing.cpu_count()

# Context window length.
context_size = 10

# Downsample setting for frequent words.
# 0 - 1e-5 is good for this
downsampling = 1e-3

# Seed for the random number generator, to make the results reproducible.
# deterministic, good for debugging
seed = 1
```

```python
lex2vec = w2v.Word2Vec(
    sg=0,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling
)
```

#### Construindo o vocabulário

```python
lex2vec.build_vocab(sentences)
```

```python
print("Word2Vec vocabulary length:", len(lex2vec.wv.vocab))
```

#### Iniciando o treinamento

```python
lex2vec.train(sentences)
```

#### Salvando modelo em arquivo para uso posterior

```python
if not os.path.exists("save"):
    os.makedirs("save")

model_name = "{}features_{}minwords_{}context.w2v".format(num_features, min_word_count, context_size)
lex2vec.save(os.path.join("save", model_name))
```

**Dando zoom em algumas áreas interessantes**

```python
def plot_region(x_bounds, y_bounds):
    slice = points[
        (x_bounds[0] <= points.x) &
        (points.x <= x_bounds[1]) & 
        (y_bounds[0] <= points.y) &
        (points.y <= y_bounds[1])
    ]
    
    fig, ax = plt.subplots(figsize=(10, 6))   
    ax = slice.plot.scatter("x", "y", s=35, ax=ax)
    for i, point in slice.iterrows():
        ax.text(point.x + 0.005, point.y + 0.005, point.word, fontsize=11)
    return fig
```

**Nomes de órgãos e empresas públicas aparecem juntos**

#### Relações lineares entre pares de palavras

```python
def nearest_similarity_cosmul(start1, end1, end2):
    similarities = lex2vec.most_similar_cosmul(
        positive=[end2, start1],
        negative=[end1]
    )
    start2 = similarities[0][0]
    print("{start1} is related to {end1}, as {start2} is related to {end2}".format(**locals()))
    #return start2
```

```python
nearest_similarity_cosmul("deputado", "camara", "senado")
nearest_similarity_cosmul("Lula", "Presidencia", "Senado")
nearest_similarity_cosmul("Aracaju", "Sergipe", "Ceara")
```
